```python
# 用matplotlib画并列柱状图
def draw_parallel_bar(x, data: list, title):
    sz = len(data)
    data = np.array(data)
    metrics = ['our model','no attention','single deep classifier']
    # years = ['2019', '2020', '2021']
    years = x
    x = np.arange(sz)
    total_width, n = 0.8, 4
    width = total_width / sz / 2
    x = x - (total_width - width) / 2

    for i in range(3): 
        if i == 1:
            plt.bar(x + i * width, data[:, i], width=width, label=metrics[i], tick_label=years)
        else :
            plt.bar(x + i * width, data[:, i], width=width, label=metrics[i])

    # plt.xlabel('metrics')
    # plt.ylabel('')
    plt.legend(title=title)
    plt.show()

data = [[0.8, 0.98, 0.82], [0.67, 0.77, 0.69], [0.64, 0.74, 0.65]]
draw_parallel_bar(['precision', 'recall', 'f1-score'], data, 'performance evaluation')
```
